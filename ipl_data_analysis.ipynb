```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("matches.csv")
```

```python
data.head()
```

```python
data.shape
```

```python
data['player_of_match'].value_counts()
```

```python
data['player_of_match'].value_counts()[0:10]
```

```python
list(data['player_of_match'].value_counts()[0:10].keys())
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(data['player_of_match'].value_counts()[0:10].keys()),list(data['player_of_match'].value_counts()[0:10]),color='g')
plt.show()
```

```python
data['result'].value_counts()
```

```python
data['toss_winner'].value_counts()
```

```python
bat_first=data[data['win_by_runs']!=0]
```

```python
bat_first.head()
```

```python
#making a histogram
plt.figure(figsize=(5,7))
plt.hist(bat_first['win_by_runs'])
plt.title("DItribution of runs")
plt.show()
```

```python
bat_first['winner'].value_counts()
```

```python
plt.figure(figsize=(6,6))
plt.bar(list(bat_first['winner'].value_counts()[0:3].keys()),list(bat_first['winner'].value_counts()[0:3]),color=["blue","yellow","orange"])
plt.show()
```

```python
#making pie chart
plt.figure(figsize=(7,7))
plt.pie(list(bat_first['winner'].value_counts()),labels=list(bat_first['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```

```python
bat_second=data[data['win_by_wickets']!=0]
```

```python
#histogram
plt.figure(figsize=(5,5))
plt.hist(bat_second['win_by_wickets'],bins=30)
plt.show()
```

```python
plt.figure(figsize=(7,7))
plt.bar(list(bat_second['winner'].value_counts()[0:3].keys()),list(bat_second['winner'].value_counts()[0:3]),color=["purple","blue","red"])
plt.show()
```

```python
#piechart
plt.figure(figsize=(5,5))
plt.pie(list(bat_second['winner'].value_counts()),labels=list(bat_second['winner'].value_counts().keys()),autopct='%0.1f%%')
plt.show()
```
